```python
import pandas as pd
from pathlib import Path
```

```python
# 1. Define Paths Robustly (Works in both Jupyter and standard scripts)
current_dir = Path.cwd()

# Handle path depending on whether you run this from root or inside the 'scripts' folder
if current_dir.name == "scripts":
    root_dir = current_dir.parent
else:
    root_dir = current_dir

seller_data_dir = root_dir / 'data' / 'csvs'
seller_name = 'StarTech'
data_file = f"{seller_name}_products.csv"
data_file_path = seller_data_dir / data_file


# Upload directory
upload_dir = root_dir / 'data' / 'csvs' / 'uploads'
```

```python
df = pd.read_csv(data_file_path)

# Remove any 'Unnamed' columns caused by trailing commas in the CSV
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# # Drop columns that are 100% empty
# df = df.dropna(axis=1, how='all')

print(f"Loaded {data_file_path} - Shape: {df.shape}")
```

```python
# Show the dataframe info
print(df.info())
```

```python
# keep='first' retains the first instance it finds and drops subsequent duplicates
df = df.drop_duplicates(subset=['seller_product_url'], keep='first')
print(f"\nTotal unique products after dropping duplicate links: {len(df)}")
```

```python
# Drop rows where 'product_name' is missing, as it's a critical field for analysis
df = df.dropna(subset=['product_name'])
print(f"\nTotal products after dropping rows with missing product names: {len(df)}")
print(df.info())
```

```python
# Remove '৳' and ',' symbol from 'current_price' and 'original_price'
df['current_price'] = df['current_price'].str.replace('৳', '').str.replace(',', '')
df['original_price'] = df['original_price'].str.replace('৳', '').str.replace(',', '')

# Convert 'current_price' and 'original_price' to numeric in fault-tolerant way (non-convertible values become NaN)
df['current_price'] = pd.to_numeric(df['current_price'], errors='coerce')
df['original_price'] = pd.to_numeric(df['original_price'], errors='coerce')

# Show the info after cleaning price columns to verify changes
print("\nAfter cleaning price columns:")
print(df.info())
```

```python
# Load the new crawl named 'StarTech_products_fixed.csv' and check the info()
fixed_currnet_price_data_file = f"{seller_name}_products_fixed.csv"
fixed_data_file_path = seller_data_dir / fixed_currnet_price_data_file
df_fixed = pd.read_csv(fixed_data_file_path)
print(f"\nLoaded {fixed_data_file_path} - Shape: {df_fixed.shape}")
print(df_fixed.info())
```

```python
# Concat the cleaned dataframe with the fixed dataframe to create a final cleaned dataframe. Make sure to drop duplicates on seller_product_url by keeping the second instance (the fixed one) since it has the correct current_price and original_price values.
final_df = pd.concat([df, df_fixed], ignore_index=True)
final_df = final_df.drop_duplicates(subset=['seller_product_url'], keep='last')

df = final_df

print(f"\nFinal cleaned dataframe shape: {df.shape}")
print(df.info())
```

```python
# Convert original_price and current_price NaN values to 0
df['current_price'] = df['current_price'].fillna(0)
df['original_price'] = df['original_price'].fillna(0)
```

```python
# Check the number of unique values in base_url
print("\nUnique base URLs:")
print(df['base_url'].nunique())
print(df['base_url'].value_counts())
```

```python
# Remove the last element after '/' in the 'base_url' column to get the main domain: https://www.startech.com.bd/camera  --> https://www.startech.com.bd
df['base_url'] = df['base_url'].apply(lambda x: x.rsplit('/', 1)[0] if isinstance(x, str) else x)
print("\nBase URLs after cleaning:")
print(df['base_url'].nunique())
print(df['base_url'].value_counts())

```

```python
# Make all the seller_slug to star-tech
df['seller_slug'] = 'star-tech'
print("\nSeller slugs after standardization:")
print(df['seller_slug'].value_counts())
```

```python
# Slugify the category_name column to populate the category_slug column

# Check how many unique category names we have
print("\nUnique category names:")
print(df['category_name'].nunique())
print(df['category_name'].value_counts())
```

```python
# Any category_name that are greater than 18 characters will be changed to 'Accessories'
df['category_name'] = df['category_name'].apply(lambda x: 'Accessories' if len(x) > 18 else x)
print("\nCategory names after standardization:")
print(df['category_name'].nunique())
print(df['category_name'].value_counts())
```

```python
# Populate category_slug by slugifying category_name
df['category_slug'] = df['category_name'].str.lower().str.replace(' ', '-').str.replace('&', 'and')
print("\nCategory slugs after slugification:")
print(df['category_slug'].nunique())
print(df['category_slug'].value_counts())
```

```python
# Populate product_slug by using the seller_product_url (last part of the URL after the last '/' removing any query parameters)
df['product_slug'] = df['seller_product_url'].apply(lambda x: x.rsplit('/', 1)[-1].split('?')[0] if isinstance(x, str) else x)
print("\nProduct slugs after extraction from URL:")
print(df['product_slug'].nunique())
print(df['product_slug'].value_counts())
```

```python
# Check the data after all cleaning steps
print("\nData after cleaning:")
print(df.info())
```

```python
# Make the first word of the product_name as brand_name and populate the brand_name column
df['brand_name'] = df['product_name'].apply(lambda x: x.split()[0] if isinstance(x, str) else x)
print("\nBrand names extracted from product names:")
print(df['brand_name'].nunique())
print(df['brand_name'].value_counts())

# The brand slug will be the same as brand_name but in lowercase
df['brand_slug'] = df['brand_name'].str.lower()
print("\nBrand slugs after conversion to lowercase:")
print(df['brand_slug'].nunique())
print(df['brand_slug'].value_counts())
```

```python
# Check the number of 'in_stock' unique values
print("\nUnique 'in_stock' values:")
print(df['in_stock'].nunique())
print(df['in_stock'].value_counts())
```

```python
# Check the review_count column to see if there are any non-numeric values that need to be cleaned
print("\nReview count unique values:")
print(df['review_count'].nunique())
print(df['review_count'].value_counts())
```

```python
# Clean the revie_count column by converting "Reviews (22)" type to "22" and also convert it to integer
df['review_count'] = df['review_count'].str.extract('(\d+)')
df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce').fillna(0).astype(int)
print("\nReview count after cleaning:")
print(df['review_count'].nunique())
print(df['review_count'].value_counts())
```

```python
# Check the df info to see if there are any columns with missing values or incorrect data types
print("\nDataframe info after all cleaning steps:")
print(df.info())
```

```python
# Save the csv file to the csvs folder with the name 'StarTech_products_cleaned.csv'
cleaned_data_file = f"{seller_name}_products_cleaned.csv"
cleaned_data_file_path = seller_data_dir / cleaned_data_file
df.to_csv(cleaned_data_file_path, index=False)
print(f"\nCleaned data saved to {cleaned_data_file_path}")
```
